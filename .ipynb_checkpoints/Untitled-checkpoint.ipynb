```python
import tensorflow as tf
import os
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import numpy as np
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2'


def add_layer(inputs, in_size, out_size, activation_function=None):
    Weights = tf.Variable(tf.random_normal([in_size, out_size]))
    biases = tf.Variable(tf.zeros([1, out_size]) + 0.1)
    Wx_plus_b = tf.matmul(inputs, Weights) + biases

    if activation_function is None:  # 即是线性关系
        outputs = Wx_plus_b
    else:
        outputs = activation_function(Wx_plus_b)
    return outputs


# data
'''
读取数据集
'''
data = pd.read_csv(open('Credit.csv'))
# data = data.values
# data.head(5)


# 将string型离散特征转化成整型
label_encoder = LabelEncoder()
data['new_Gender'] = label_encoder.fit_transform(data['Gender'])    # 男 0;女 1
data['new_Student'] = label_encoder.fit_transform(data['Student'])  #  Yes 1; No 0
data['new_Married'] = label_encoder.fit_transform(data['Married'])  #  Yes 1. No 0
data['new_Ethnicity'] = label_encoder.fit_transform(data['Ethnicity'])  #  Caucasian 2; Asian 1; African American 0
# data.head(5)


# 重新选取转化后的字段为新数据集
new_data = data[['Income', 'Limit', 'Rating', 'Cards', 'Age', 'Education',
                 'new_Gender', 'new_Student', 'new_Married', 'new_Ethnicity', 'Balance']]

X_data = new_data[['Income', 'Limit', 'Rating', 'Cards', 'Age', 'Education',
                 'new_Gender', 'new_Student', 'new_Married', 'Balance']]

Y_data = new_data[['Balance']]

```

```python
import numpy as np
import urllib.request as urllib
from sklearn.preprocessing import StandardScaler,MinMaxScaler,LabelEncoder,OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from keras.models import Sequential
from keras.layers import Dense
```

```python
prediction_network = Sequential()
prediction_network.add(Dense(units=10, kernel_initializer='uniform', activation='relu', input_dim=10))
prediction_network.add(Dense(units=10, kernel_initializer='uniform', activation = 'relu'))
prediction_network.add(Dense(units=1, kernel_initializer='uniform'))
prediction_network.compile(optimizer='adam', loss='mean_squared_error', metrics=['accuracy'])

# Fitting model
prediction_network.fit(X_data, Y_data, batch_size=10, epochs=100)

# Predict the test set
y_pred = prediction_network.predict(X_data)
rmse = mean_squared_error(Y_data, y_pred)
print("Root Mean Square Error : {:.4f}".format(rmse))
```
